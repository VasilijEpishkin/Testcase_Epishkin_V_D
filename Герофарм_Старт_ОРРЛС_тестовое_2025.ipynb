Используя подготовленный на прошлом шаге датасет, постройте графики, которые будут отражать степень сходства (или различия) GPX-007 с молекулами из патентной заявки.  
Вы можете выбрать те варианты визуализации данных, которые сочтете наиболее подходящими. Не забудьте сохранить код для построения графиков в ячейке ниже, чтобы на этапе проверки изображения были воспроизведены корректно.

```python
!pip install Levenshtein
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle, Patch
import Levenshtein
from io import StringIO

df['Moiety_Letter'] = df['branch_site'].apply(lambda x: x.split('Moiety ')[1].replace(')', ''))

align_map = {
    '[2meA]': 'B', '[Aib]': 'B',
    '[2meL]': 'J', '[αMeL]': 'J', '[aMeL]': 'J',
    '[norleucine]': 'O', '[norvaline]': 'U'
}

def tokenize(seq):
    """Tokenize sequence into single AA or [bracketed] groups"""
    seq = seq.replace(" ", "")
    tokens = []
    i = 0
    while i < len(seq):
        if seq[i] == '[':
            end = seq.find(']', i)
            tokens.append(seq[i:end+1])
            i = end + 1
        else:
            tokens.append(seq[i])
            i += 1
    return tokens

def to_align_string(tokens):
    """Convert tokens to alignment-friendly string"""
    return "".join([align_map.get(t, t) for t in tokens])

def align_to_ref(target_seq, ref_seq_str):
    """Align target sequence to reference using Levenshtein distance"""
    tgt_tokens = tokenize(target_seq)
    tgt_str = to_align_string(tgt_tokens)

    ops = Levenshtein.editops(ref_seq_str, tgt_str)
    aligned = list(tgt_tokens)

    for op_type, s_i, d_i in ops:
        if op_type == 'delete':
            if d_i <= len(aligned):
                if d_i == len(aligned) or aligned[d_i] != '-':
                    aligned.insert(d_i, '-')
    return aligned[:45]

ref_tokens = tokenize(df.iloc[1]['sequence'])
ref_str = to_align_string(ref_tokens)
aligned_data = [align_to_ref(row['sequence'], ref_str) for idx, row in df.iterrows()]

fig, ax = plt.subplots(figsize=(26, 10))

colors_aa = {
    'A': '#c8e6c9', 'G': '#c8e6c9', 'V': '#c8e6c9', 'L': '#c8e6c9', 'I': '#c8e6c9',
    'F': '#e1f5fe', 'W': '#e1f5fe', 'Y': '#e1f5fe',
    'K': '#ffccbc', 'R': '#ffccbc', 'H': '#ffccbc',
    'D': '#ffe0b2', 'E': '#ffe0b2',
    'S': '#fff9c4', 'T': '#fff9c4', 'N': '#fff9c4', 'Q': '#fff9c4', 'C': '#fff9c4',
    'P': '#f5f5f5',
    '-': '#ffffff'
}
special_color = '#e1bee7'

colors_moiety = {
    'G': '#D62728',
    'A': '#1f77b4',
    'B': '#2ca02c',
    'C': '#ff7f0e',
    'D': '#9467bd'
}

box_h = 0.8
box_w = 1.0
start_y = len(df) - 1

for row_idx, tokens in enumerate(aligned_data):
    y = start_y - row_idx
    row_info = df.iloc[row_idx]
    mol_id = row_info['id']
    moiety_let = row_info['Moiety_Letter']

    m_color = colors_moiety.get(moiety_let, 'gray')
    ax.add_patch(Rectangle((-3.0, y), 2.0, box_h, facecolor=m_color, edgecolor='black', lw=1))
    ax.text(-2.0, y + box_h/2, f"Moiety {moiety_let}",
            color='white', fontweight='bold', ha='center', va='center', fontsize=9)

    font_weight = 'bold' if mol_id in ['GPX-007', 'Retatrutide'] else 'normal'
    color_label = 'red' if mol_id == 'GPX-007' else 'blue' if mol_id == 'Retatrutide' else 'black'

    ax.text(-3.3, y + box_h/2, mol_id,
            ha='right', va='center',
            fontweight=font_weight, fontsize=11, color=color_label, family='monospace')

    for col_idx, aa in enumerate(tokens):
        clean_name = aa.replace('[','').replace(']','')

        if '[' in aa:
            fill = special_color
        else:
            fill = colors_aa.get(clean_name, '#ffffff')

        if aa == '-':
            ax.plot([col_idx+0.2, col_idx+0.8], [y+box_h/2, y+box_h/2],
                   color='gray', linewidth=1.5)
            continue

        rect = Rectangle((col_idx, y), box_w, box_h,
                        facecolor=fill, edgecolor='gray', linewidth=0.5)
        ax.add_patch(rect)

        f_size = 8 if len(clean_name) > 3 else 10 if len(clean_name) > 1 else 11
        ax.text(col_idx + 0.5, y + 0.5*box_h, clean_name,
                ha='center', va='center', fontsize=f_size,
                fontweight='bold', color='#333333')

    if mol_id == 'GPX-007':
        rect = Rectangle((-3.1, y-0.05), len(tokens)+3.2, box_h+0.1,
                        fill=False, edgecolor='red', linewidth=2.5, zorder=5)
        ax.add_patch(rect)
    elif mol_id == 'Retatrutide':
        rect = Rectangle((-3.1, y-0.05), len(tokens)+3.2, box_h+0.1,
                        fill=False, edgecolor='blue', linewidth=2,
                        linestyle='--', zorder=5)
        ax.add_patch(rect)

ax.set_xlim(-5.5, 43)
ax.set_ylim(-0.5, len(df))
ax.set_yticks([])
ax.set_xlabel("Aligned Sequence Position", fontweight='bold', fontsize=12)
ax.xaxis.tick_bottom()

ax.set_title("Structural Alignment",
             fontsize=16, fontweight='bold', pad=20)

legend_elements = [
    # Amino acid categories
    Patch(facecolor=colors_aa['L'], edgecolor='gray', label='Aliphatic (L, I, V, A, G)'),
    Patch(facecolor=colors_aa['Y'], edgecolor='gray', label='Aromatic (F, Y, W)'),
    Patch(facecolor=colors_aa['K'], edgecolor='gray', label='Positive (K, R, H)'),
    Patch(facecolor=colors_aa['D'], edgecolor='gray', label='Negative (D, E)'),
    Patch(facecolor=colors_aa['S'], edgecolor='gray', label='Polar (S, T, N, Q, C)'),
    Patch(facecolor=special_color, edgecolor='gray', label='Non-canonical AA'),
    Patch(facecolor='white', edgecolor='gray', label='Gap/Deletion'),
]

ax.legend(handles=legend_elements,
         loc='upper center',
         bbox_to_anchor=(0.5, -0.05),
         ncol=6,
         frameon=False,
         fontsize=10)

plt.tight_layout()
plt.savefig('/content/Testcase_Epishkin_V_D/alignment.png', dpi=300, bbox_inches='tight')
plt.show()
```

## **Задание №4.**  

Пока вы принимали решение о дальнейшей разработке молекулы, коллеги из лаборатории пептидного синтеза уже приступили к подготовке оборудования и реагентов. Для закупки корректного количества расходных материалов они запросили у вас следующие характеристики молекулы GPX-007:  
*   molecular_weight_Da - молекулярная масса в Да  
*   S_S_bond_count - количество дисульфидных связей
*   aromatic_count - количество ароматических аминокислотных остатков
*   empiric_formula - формула соединения в формате CvHwNxOySz, где v, w, x, y, z - количество атомов углерода, водорода, азота, кислорода и серы, соотвественно.

 Рассчитайте эти параметры для молекулы GPX-007 и приведите ответ в ячейке ниже. При использовании python приведите код, при использовании сторонних инструментов - кратко опишите алгоритм, по которому производили расчеты.  

```python
import pandas as pd

def calculate_gpx007_descriptors(df_data, target_id="GPX-007"):
    sequence_raw = df_data[df_data['id'] == target_id]['sequence'].values[0]

    aa_formulas = {
        'A': {'C': 3, 'H': 7, 'N': 1, 'O': 2, 'S': 0},
        'D': {'C': 4, 'H': 7, 'N': 1, 'O': 4, 'S': 0},
        'E': {'C': 5, 'H': 9, 'N': 1, 'O': 4, 'S': 0},
        'F': {'C': 9, 'H': 11, 'N': 1, 'O': 2, 'S': 0},
        'G': {'C': 2, 'H': 5, 'N': 1, 'O': 2, 'S': 0},
        'I': {'C': 6, 'H': 13, 'N': 1, 'O': 2, 'S': 0},
        'K': {'C': 6, 'H': 14, 'N': 2, 'O': 2, 'S': 0},
        'L': {'C': 6, 'H': 13, 'N': 1, 'O': 2, 'S': 0},
        'P': {'C': 5, 'H': 9, 'N': 1, 'O': 2, 'S': 0},
        'Q': {'C': 5, 'H': 10, 'N': 2, 'O': 3, 'S': 0},
        'S': {'C': 3, 'H': 7, 'N': 1, 'O': 3, 'S': 0},
        'T': {'C': 4, 'H': 9, 'N': 1, 'O': 3, 'S': 0},
        'Y': {'C': 9, 'H': 11, 'N': 1, 'O': 3, 'S': 0},
    }

    unnatural_aa = {
        '[2meA]': {'C': 4, 'H': 9, 'N': 1, 'O': 2, 'S': 0},
        '[2meL]': {'C': 7, 'H': 15, 'N': 1, 'O': 2, 'S': 0},
    }

    # Компоненты Moiety G
    # Структура: Lys(epsilon) -> [AEEA] -> [AEEA] -> [gamma-Glu] -> [C20-diacid]
    fatty_acid_C20 = {'C': 20, 'H': 38, 'N': 0, 'O': 4, 'S': 0}
    gamma_Glu      = {'C': 5,  'H': 9,  'N': 1, 'O': 4, 'S': 0}
    AEEA_linker    = {'C': 6,  'H': 13, 'N': 1, 'O': 4, 'S': 0}

    atomic_weights = {'C': 12.011, 'H': 1.008, 'N': 14.007, 'O': 15.999, 'S': 32.06}

    residues = []
    i = 0
    while i < len(sequence_raw):
        if sequence_raw[i] == '[':
            end = sequence_raw.find(']', i)
            residues.append(sequence_raw[i:end+1])
            i = end + 1
        else:
            residues.append(sequence_raw[i])
            i += 1

    total_atoms = {'C': 0, 'H': 0, 'N': 0, 'O': 0, 'S': 0}

    for res in residues:
        if res in aa_formulas:
            form = aa_formulas[res]
        elif res in unnatural_aa:
            form = unnatural_aa[res]
        else:
            raise ValueError(f"Unknown residue: {res}")

        for atom, count in form.items():
            total_atoms[atom] += count

    # 1 Fatty Acid + 1 gammaGlu + 2 AEEA
    components_moiety = [fatty_acid_C20, gamma_Glu, AEEA_linker, AEEA_linker]
    for comp in components_moiety:
        for atom, count in comp.items():
            total_atoms[atom] += count

    # Backbone N остатков -> N-1 связей
    backbone_bonds = len(residues) - 1

    # Moiety G bonds
    # 1. Lys(eps)-NH2 + HOOC-AEEA -> Amide bond (1)
    # 2. AEEA-NH2 + HOOC-AEEA -> Amide bond (1)
    # 3. AEEA-NH2 + HOOC(gamma)-Glu -> Amide bond (1)
    # 4. Glu-NH2 + HOOC-FattyAcid -> Amide bond (1)
    sidechain_bonds = 4

    total_water_lost = backbone_bonds + sidechain_bonds

    total_atoms['H'] -= 2 * total_water_lost
    total_atoms['O'] -= 1 * total_water_lost
    total_atoms['N'] += 1
    total_atoms['H'] += 1
    total_atoms['O'] -= 1

    mw = sum(total_atoms[a] * atomic_weights[a] for a in total_atoms)

    aromatic_residues = ['F', 'Y', 'W', 'H']
    aromatic_count = sum(1 for res in residues if res in aromatic_residues)

    cys_count = residues.count('C')
    ss_bond_count = cys_count // 2

    formula_str = f"C{total_atoms['C']}H{total_atoms['H']}N{total_atoms['N']}O{total_atoms['O']}S{total_atoms['S']}"

    return mw, ss_bond_count, aromatic_count, formula_str, len(residues)

mw, ss, aro, form, length = calculate_gpx007_descriptors(df, target_id="GPX-007")

print(f"molecular_weight_Da = {mw:.2f}")
print(f"S_S_bond_count = {ss}")
print(f"aromatic_count = {aro}")
print(f"empiric_formula = {form}")
```

Для GPX‑007 на основе реализованного скрипта получены следующие параметры:

molecular_weight_Da ≈ 4890.6

S_S_bond_count = 0 (в аминокислотной последовательности отсутствуют остатки Cys, поэтому дисульфидные мосты не формируются)

aromatic_count = 5 (учитывались ароматические аминокислоты F, Y и W)

empiric_formula = C228H355N47O71S0

Расчёт массы и эмпирической формулы выполнялся путём суммирования атомных составов отдельных аминокислотных остатков с учётом потери воды при образовании пептидных связей; непротеиногенные фрагменты заданы через отдельный словарь с их формулами.
